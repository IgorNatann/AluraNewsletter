# Aula 1 - Consultando o banco de dados com um assistente

Usamos a biblioteca [sqlalchemy](https://docs.sqlalchemy.org/en/20/) para trabalhar com bancos de dados no ambiente Python.
- O [`create_engine`](https://docs.sqlalchemy.org/en/20/core/engines.html) cria a conexão com o banco de dados usando a URL fornecida.
- O [`MetaData`](https://docs.sqlalchemy.org/en/20/core/metadata.html) gerencia informações sobre as tabelas e seus esquemas (metadados).

````python
from sqlalchemy import create_engine, MetaData //biblioteca para conexão com o banco de dados

url = '/content/ecommerce.db' //caminho do banco de dados
engine = create_engine(f'sqlite:///{url}') //cria uma string de conexão sqllite com o banco de dados
````

````python
metadata_obj = MetaData() //cria um objeto para armazenar informações sobre os metadados do banco de dados
metadata_obj.reflect(engine) //reflectar os metadados do banco de dados no objeto metadata_obj
metadata_obj.tables.keys() //retorna o nome das tabelas no banco de dados
````

* [`pd.read_sql_table`](https://pandas.pydata.org/docs/reference/api/pandas.read_sql_table.html) carrega a tabela diretamente para um DataFrame do pandas

````python
import pandas as pd //biblioteca para manipulação de dados

pd.read_sql_table('Clientes', engine) //carrega a tabela Clientes para um DataFrame do pandas
````

````python
pd.read_sql_table('Fornecedores', engine) //carrega a tabela Fornecedores para um DataFrame do pandas
````

````python
pd.read_sql_table('Funcionarios', engine) //carrega a tabela Funcionarios para um DataFrame do pandas
````
